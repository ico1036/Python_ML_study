## Estimate pi using MonteCarlo algorithm

```python
import random
```

```python
n=10000  # Number of iteration: You can modify it other number
count =0 # N of in-circle

# ----- arrays for visualization
arr_incircle_x = []
arr_incircle_y = []
arr_outcircle_x = []
arr_outcircle_y = []
```

```python
count=0
for i in range(n):
    
    # -- Uniform random varaible from 0 to 1
    x=random.uniform(0,1.0)
    y=random.uniform(0,1.0)
    
    # -- In-circle
    if (x**2+y**2)<=1:
        arr_incircle_x.append(x)
        arr_incircle_y.append(y)
        count=count+1
    else:
        arr_outcircle_x.append(x)
        arr_outcircle_y.append(y)
    
    # -- Calculate Pi
    if count % 1000 == 0:
        print("Pi? ",4*count/n)
```

```python
# Import Library
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rc('xtick',labelsize=15)
plt.rc('ytick',labelsize=15)
plt.rcParams["figure.figsize"] = (8,8)

theta = np.linspace(0, 2*np.pi, 100)
r = np.sqrt(1.0)
x1 = r*np.cos(theta)
x2 = r*np.sin(theta)
```

```python
plt.plot(x1,x2,linewidth=4,color='r')
plt.scatter(arr_incircle_x,arr_incircle_y,color='salmon',marker='.',Label="In-circle")
plt.scatter(arr_outcircle_x,arr_outcircle_y,color='deepskyblue',marker='.',Label="Out-circle")
plt.legend()
plt.xlim([0,1])
plt.ylim([0,1])
plt.title("Test {0} number of MC events".format(n),size=20)
plt.legend(loc="lower left",prop={'size':15})
```
